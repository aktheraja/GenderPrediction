```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
```

```python
img = cv2.imread('IMG_0239.jpg', cv2.IMREAD_GRAYSCALE)
# IMREAD_GRAYSCALE, 0
# IMREAD_COLOR, 1
# IMREAD_UNCHANGED, -1
```

```python
plt.imshow(img, cmap='gray', interpolation='bicubic')
```

```python
resized = cv2.resize(img, (100,100))
```

```python
plt.imshow(resized, cmap='gray', interpolation='bicubic')
```

#### Streaming from webcam

Notebook is glitchy with this and may require restarting kernel after closing the videoframe

```python
cap = cv2.VideoCapture(0)  # 0 for the first webcam in the system
fourcc = cv2.VideoWriter_fourcc(*'XVID')
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640,480))

while True:
    ret, frame = cap.read()  # if there is true, ret would get True value
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    out.write(frame) # write output to file
    
    cv2.imshow('frame', gray)
    
    k = cv2.waitKey(30) & 0xff  # ESC key to stop
    if k == 27:
        break

cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
